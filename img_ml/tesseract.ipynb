```python
import cv2
import numpy as np
import pytesseract
import re
import matplotlib.pyplot as plt
```

```python
image  = cv2.imread('test2.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# 이미지를 표시
plt.figure(figsize=(5, 5))
plt.imshow(gray, cmap='gray')
plt.title('Gray Scaled Image')
plt.show()
```

```python
# 노이즈 제거를 위한 가우시안 블러
blur = cv2.GaussianBlur(gray, (3,3), 0)

# 블러된 이미지를 표시
plt.figure(figsize=(5, 5))
plt.imshow(blur, cmap='gray')
plt.title('Blurred Image')
plt.show()
```

```python
# 이진화
ret, threshed = cv2.threshold(blur, 127, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

# 이진화된 이미지를 표시
plt.figure(figsize=(5, 5))
plt.imshow(threshed, cmap='gray')
plt.title('Thresholded Image')
plt.show()
```

```python
# pytesseract를 이용한 OCR
custom_config = r'--oem 3 --psm 6'
result = pytesseract.image_to_string(threshed, config=custom_config)
```

```python
print(result)
```

```python
# 결과에서 년-월-일-시-분-초 형식만 추출
date_time = re.findall(r'\d{4}.\d{2}.\d{2}', result)
for dt in date_time:
    print(dt)
```
