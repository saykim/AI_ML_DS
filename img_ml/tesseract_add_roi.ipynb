```python
# import dependency files
import cv2
import numpy as np
import pytesseract

# 이미지 업로드
img = cv2.imread('snack1.jpg')

# 마우스를 이용하여 이미지에서 원하는 부분을 ROI로 지정한다
x,y,w,h = cv2.selectROI('img',img,False)

if w and h:
    roi = img[y:y+h, x:x+w]
    cv2.imshow('cropped',roi)
    cv2.imwrite('test/cropped.jpg',roi)

    # tesseract를 이용해서 ocr을 진행하도록 한다
    ocr = pytesseract.image_to_string('test/cropped.jpg',lang='kor')
    print(ocr)
    
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# tesseract 일부 문자만 인식

```python
import cv2
import pytesseract
import matplotlib.pyplot as plt
from tkinter import filedialog
from tkinter import Tk

def ocr_on_image(image_path):
    img = cv2.imread(image_path)
    if img is not None:
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        blurred = cv2.GaussianBlur(gray, (3, 3), 0)
        binary = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 21, 2)
        plt.imshow(binary, cmap='gray')
        plt.show()

        # config 설정을 통해 숫자와 '.' ','만 인식하도록 함
        custom_config = r'-c tessedit_char_whitelist=0123456789. --oem 3 --psm 6'
        ocr = pytesseract.image_to_string(binary, config=custom_config)
        print(ocr)
    else:
        print("Image file not found.")

root = Tk()
root.withdraw()
file_path = filedialog.askopenfilename()
ocr_on_image(file_path)
```

```python
import cv2
import pytesseract
import matplotlib.pyplot as plt
import re
from tkinter import filedialog
from tkinter import Tk

def ocr_on_image(image_path):
    img = cv2.imread(image_path)
    if img is not None:
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        blurred = cv2.GaussianBlur(gray, (3, 3), 0)
        binary = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 21, 2)
        plt.imshow(binary, cmap='gray')
        plt.show()

        # config 설정을 통해 숫자와 '.'만 인식하도록 함
        custom_config = r'-c tessedit_char_whitelist=0123456789. --oem 3 --psm 6'
        ocr = pytesseract.image_to_string(binary, config=custom_config)

        # 정규 표현식을 사용하여 날짜 형식의 문자열만 추출
        dates = re.findall(r'\d{4}\.\d{2}\.\d{2}', ocr)
        for date in dates:
            print(date)
    else:
        print("Image file not found.")

root = Tk()
root.withdraw()
file_path = filedialog.askopenfilename()
ocr_on_image(file_path)
```

```python
import cv2
import pytesseract
from pytesseract import Output
import matplotlib.pyplot as plt
from tkinter import filedialog
from tkinter import Tk

def ocr_on_image(image_path):
    img = cv2.imread(image_path)
    if img is not None:
        # 이미지의 방향을 얻음
        osd = pytesseract.image_to_osd(img, output_type=Output.DICT)
        rotate_angle = osd['rotate']

        # 이미지를 회전시킴
        height, width = img.shape[:2]
        center = (width / 2, height / 2)
        M = cv2.getRotationMatrix2D(center, rotate_angle, 1)
        rotated = cv2.warpAffine(img, M, (width, height))

        plt.imshow(rotated)
        plt.show()

        # OCR 적용
        custom_config = r'-c tessedit_char_whitelist=0123456789. --oem 3 --psm 6'
        ocr = pytesseract.image_to_string(rotated, config=custom_config)
        print(ocr)
    else:
        print("Image file not found.")

root = Tk()
root.withdraw()
file_path = filedialog.askopenfilename()
ocr_on_image(file_path)
```

```python
import cv2
import pytesseract
from pytesseract import Output
import imutils

# 이미지 경로
image_path = "test2.png"

# 이미지 로드
image = cv2.imread(image_path)

# OSD 모드에서 테서렉트 실행
osd = pytesseract.image_to_osd(image, output_type=Output.DICT)

# 회전 각도를 얻습니다.
rotate_angle = osd['rotate']

# 이미지를 회전하여 텍스트 방향을 수정합니다.
rotated = imutils.rotate_bound(image, angle=rotate_angle)

# 수정된 이미지에 OCR을 적용하고 결과를 출력합니다.
print(pytesseract.image_to_string(rotated))
```

```python
import cv2
import imutils
import pytesseract
from PIL import Image

def increase_dpi(image_path, output_path, dpi):
    # Open an image file
    im = Image.open(image_path)
    # Save the image with a new dpi
    im.save(output_path, dpi=(dpi, dpi))

def correct_orientation(image_path):
    # Load the image from disk
    image = cv2.imread(image_path)
    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # Use Tesseract to detect the orientation of the text
    osd = pytesseract.image_to_osd(gray)
    # Parse the OSD output
    angle = osd.split('\n')[1].split(': ')[1]
    # Rotate the image to correct the orientation
    rotated = imutils.rotate_bound(image, angle=-int(angle))
    return rotated

def main():
    # Path of the image file
    image_path = 'snack1.jpg'
    # Path where the output image will be saved
    output_path = 'output.jpg'
    # Increase the dpi of the image
    increase_dpi(image_path, output_path, 300)
    # Correct the orientation of the text in the image
    corrected_image = correct_orientation(output_path)
    # Display the corrected image
    cv2.imshow("Corrected Image", corrected_image)
    cv2.waitKey(0)

if __name__ == "__main__":
    main()
```
