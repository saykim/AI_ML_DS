```python
path = '/content/drive/MyDrive/Colab Notebooks/10.cj_project/datasets/img_cls/fm_img_hori'
```

```python
def prepare_dataset(image_folder):
    data = []
    for label in ['good', 'ng']:
        class_folder = os.path.join(image_folder, label)
        for image_file in os.listdir(class_folder):
            data.append({'image': os.path.join(class_folder, image_file), 'label': label})
    return pd.DataFrame(data)

image_folder = path
dataset = prepare_dataset(image_folder)

dataset
```
